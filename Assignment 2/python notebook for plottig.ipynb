```python
ds=pd.read_csv("Question_1.csv")

f= open("Question_1.txt", "r")
T=(f.read())
print(type(T))
time=float(T)
# print(time)
```

```python
close()
figure(1)
title("Acoustic Equation")
title("Pressure VS X Initial Condition ")
plot(ds.Location,ds.init_P)
xlabel("x") 
ylabel("P")
# savefig('Initial Condition.png')
figure(2)
grid("ON")
title("Pressure Vs X\n at Time step ={} sec" .format(round(time,3)))
xlabel("x") 
ylabel("P")
plot(ds.Location,ds.P_1st)
plot(ds.Location,ds.P_2nd)

legend(["First order","Second order"])
savefig('Presuure VS Location at time step {}.png'.format(round(time,3)))
figure(3)
grid("ON")
# title(X)
title("Velocity Vs X\n at Time step ={} sec" .format(round(time,3)))
plot(ds.Location,ds.U_1st)
plot(ds.Location,ds.U_2nd)
legend(["First order","Second order"])
xlabel("x") 
ylabel("U")
savefig('Velocity VS Location at time step {}.png'.format(round(time,3)))

show()
```
